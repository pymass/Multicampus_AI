```python
x = np.arange(-10,11,1)
f_x = x ** 2
```

```python
plt.plot(x, f_x)
plt.show()
```

```python
x_new = 10
derivative = []
y = []
learning_rate = 0.1
for i in range(100):
    old_value = x_new
    derivative.append(old_value - learning_rate * 2 * old_value)
    x_new = old_value - learning_rate * 2 * old_value
    y.append(x_new ** 2)
```

```python
plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```

```python
def sin_function(x):
    return x * np.sin(x ** 2) + 1
def derivitive_f(x):
    return np.sin(x ** 2) + 2 * (x**2) * np.cos(x ** 2)
```

```python
x = np.arange(-3,3,0.001)
f_x = sin_function(x)
```

```python
plt.plot(x, f_x)
plt.show()
```

```python
derivitive_f(3)
```

```python
x_new = 1 
derivative = [] # 
y = [] 
learng_rate= 0.01 
for i in range(10000): 
    old_value = x_new 
    x_new = old_value - learng_rate * derivitive_f(old_value) 
    derivative.append(x_new) 
    y.append(sin_function(x_new))
plt.plot(x, f_x) 
plt.scatter(derivative, y) 
plt.show()
```

```python
from sklearn.datasets import load_boston 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
boston = load_boston()
boston['data']
```

```python
x_data = boston.data
y_data = boston.target.reshape(boston.target.size,1)
```

```python
y_data.shape
```

```python
from sklearn import preprocessing

minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,5)).fit(x_data) 
x_scaled_data = minmax_scale.transform(x_data)
x_scaled_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_scaled_data, y_data, test_size=0.33)
# training set test set 
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=8)
# fit_intercept , normalize 
regr.fit(X_train, y_train)
regr
```

```python
regr.coef_ , regr.intercept_
```

```python
print('Coefficients: ', regr.coef_) # y = ax + b  a / data  13 
print('intercept: ', regr.intercept_) # y = ax + b  b
```

```python
x_data
```

```python
regr.predict(x_data[:5]) # 5   
```

```python
x_data[:5].dot(regr.coef_.T) + regr.intercept_ #  
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
y_true = y_test
y_hat = regr.predict(X_test)

r2_score(y_true, y_hat), mean_absolute_error(y_true, y_hat), mean_squared_error(y_true, y_hat)
# X_test
```

```python
y_true = y_train 
y_hat = regr.predict(X_train)
r2_score(y_true, y_hat), mean_absolute_error(y_true, y_hat), mean_squared_error(y_true, y_hat)
```
