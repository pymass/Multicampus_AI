### 문제 8
-----
sakila의 rental 테이블에서 월별 렌탈 수 출력 날짜 오름차순  
pandas, sql

```python
rental_df = pd.read_sql('select * from rental', sakila_db)
rental_df.head()
```

```python
rental_df.index
```

```python
from datetime import datetime

rental_df['rental_month'] = 0
for i in rental_df.index:
    rental_df.loc[i,'rental_month'] = rental_df['rental_date'][i].month
```

```python
rental_df
```

```python
rental_df.groupby('rental_month')['rental_id'].count().sort_values()
```

```python
sql = '''
select left(rental_date, 7) 'month',count(rental_id)
from rental group by month order by count(rental_id);
'''

pd.read_sql(sql, sakila_db)
```

### 문제 8
-----
sakila의 payment 테이블에서 월별 전체 매출, 평균 매출, 결제 갯수를 컬럼으로하는 dataframe만들기  
pandas, sql

```python
payment_df = pd.read_sql('select * from payment', sakila_db)
payment_df.head()
```

```python
payment_df.head()
```

```python
import datetime

payment_df['payment_month'] = 0
for i in payment_df.index:
    payment_df.loc[i, 'payment_month'] = payment_df['payment_date'][i].month
```

```python
month_amount = payment_df.groupby('payment_month').sum()['amount']
month_amount
```

```python
month_mean_amount = payment_df.groupby('payment_month').mean()['amount']
month_mean_amount
```

```python
count_amount = payment_df.groupby('payment_month').count()['amount']
count_amount
```

```python
import numpy as np
payment_df.groupby('payment_month').agg({'amount' : [np.mean, np.sum], 'payment_id' : np.size})
```

```python
join_df = pd.concat([month_amount, month_mean_amount, count_amount], axis=1)
join_df
```

```python
join_df.columns = ['매출액', '평균매출액', '결제갯수']
del join_df.index.name
```

```python
join_df
```

```python
sql = '''
select left(payment_date, 7) '월', sum(amount) '매출액', avg(amount) '평균매출액',
count(amount) '결제갯수' from payment group by 월
'''

pd.read_sql(sql, sakila_db)
```
