```python
sakila_db = MySQLdb.connect('localhost', 'root', '1234', 'sakila', charset='utf8')
```

```python
sql = 'show tables'
pd.read_sql(sql, sakila_db)
```

```python
sql = 'select * from rental'
rental_df = pd.read_sql(sql, sakila_db)
```

```python
sql = 'select * from customer'
customer_df = pd.read_sql(sql, sakila_db)
```

```python
rental_df.head()
```

```python
customer_df
```

```python
rental_df.columns
```

```python
rental_df['rental_date'] = rental_df['rental_date'].astype(str)
```

```python
rental_df['rental_month'] = 0
for i in rental_df.index:
    rental_df.loc[i, 'rental_month'] = rental_df['rental_date'][i][:7]
```

```python
rental_df.groupby('rental_month')['rental_id'].count()
```

```python
join_df = pd.merge(rental_df, customer_df, on='customer_id')
```

```python
join_df.head()
```

```python
pivot = join_df.pivot_table(values='active',index='customer_id',columns='rental_month', aggfunc='count')
```

```python
pivot
```

```python
sql = '''
select first_name, email, count(rental_id) num_rent
from customer c join rental r
on c.customer_id = r.customer_id
group by c.customer_id
having num_rent > 30
order by num_rent desc;
'''

pd.read_sql(sql, sakila_db)
```
