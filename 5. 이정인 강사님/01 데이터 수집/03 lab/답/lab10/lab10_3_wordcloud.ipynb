```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline 

text=open("일본_naver_news.m","r", encoding="utf-8").read()

spwords=set(STOPWORDS) ## 기본적으로 많이 쓰는 단어
spwords.add('일본') ##  제외하고 싶은 단어 추가

wordcloud = WordCloud(max_font_size=200,font_path='c:/Windows/fonts/malgun.ttf',
                      stopwords=spwords,background_color='#FFFFFF',
                      max_words = 2000,
                      colormap="spring_r",
                      width=1200,height=800).generate(text)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')

plt.savefig('japan.png', bbox_inches='tight')
plt.show()
```
