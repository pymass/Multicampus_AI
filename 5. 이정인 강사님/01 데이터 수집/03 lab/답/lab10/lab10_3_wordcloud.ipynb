```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline 

text=open("_naver_news.m","r", encoding="utf-8").read()

spwords=set(STOPWORDS) ##    
spwords.add('') ##     

wordcloud = WordCloud(max_font_size=200,font_path='c:/Windows/fonts/malgun.ttf',
                      stopwords=spwords,background_color='#FFFFFF',
                      max_words = 2000,
                      colormap="spring_r",
                      width=1200,height=800).generate(text)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')

plt.savefig('japan.png', bbox_inches='tight')
plt.show()
```
