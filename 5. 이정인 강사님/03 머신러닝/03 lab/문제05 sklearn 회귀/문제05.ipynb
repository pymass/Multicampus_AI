```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('지역_위치별(주유소).xls')
df.head()
```

```python
df.replace('-', np.NaN,inplace=True)
```

```python
X_train = df.dropna()[['휘발유', '경유']]
y_train = df.dropna()['고급휘발유']
```

```python
X_test = df.loc[df['고급휘발유'].isnull(), ['휘발유','경유']]
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
print(pred)
```

```python
pred.astype(int)
```

```python
df.loc[df['고급휘발유'].isnull(), '고급휘발유'] = pred.astype(int)
```
