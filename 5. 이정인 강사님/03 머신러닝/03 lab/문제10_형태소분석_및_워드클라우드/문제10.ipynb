```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
```

```python
import codecs
from konlpy.tag import Okt
```

```python
t1 = codecs.open("twitter.txt","r",encoding="utf-8")  
```

```python
t = open("twitter.m","w",encoding="utf-8") 

# 형태소 분석
twitter = Okt()
i = 0

while True:
    line = t1.readline()
    if not line: break

    if i % 100 == 0:
        print("current - " + str(i)) #100라인에 한번씩 출력
    i += 1

    malist = twitter.pos(line, norm=True, stem=True)

    r = []
    for word in malist:
                   # 조사/어미/구두점 등은 대상에서 제외
        #if not word[1] in ["Josa", "Eomi", "Punctuation"]:
           # t.write(word[0] + " ")
        
        
        if word[1] in ["Noun"]:
            t.write(word[0] + " ")

t.close() 
```

```python
data=open("twitter.m","r", encoding="utf-8").read()
```

```python
spwords = set(STOPWORDS)
wordcloud = WordCloud(font_path='c:/Windows/fonts/malgun.ttf', max_words=2000,
                     stopwords=spwords).generate(data)
```

```python
plt.figure(figsize=(10,9))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')
```
