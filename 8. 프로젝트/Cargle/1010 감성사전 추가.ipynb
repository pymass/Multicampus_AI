```python
import numpy as np
import pandas as pd
from konlpy.tag import Komoran
```

```python
ko = Komoran()
```

```python
sentiword = pd.read_csv('./KnuSentiLex-master/SentiWord_Dict.txt', sep='\t', header=None)
```

```python
sentiword
```

```python
sentiword[1] = np.where(sentiword[1] == 1, 0.5, sentiword[1])
sentiword[1] = np.where(sentiword[1] == 2, 1.0, sentiword[1])
sentiword[1] = np.where(sentiword[1] == -1, -0.5, sentiword[1])
sentiword[1] = np.where(sentiword[1] == -2, -1.0, sentiword[1])
```

```python
polarity = pd.read_csv('./lexicon/polarity.csv')
polarity.head()
```

```python
polarity['max.value'].unique()
```

```python
polarity = polarity.loc[(polarity['max.value'] == 'POS') | (polarity['max.value'] == 'NEG')]
```

```python
polarity['max.value'].unique()
```

```python
polarity['score'] = np.where(polarity['POS'] > polarity['NEG'], polarity['max.prop'], polarity['max.prop'] * -1)
```

```python
polarity
```

```python
sentiword.head()
```

```python
sentiword[0] = sentiword[0].apply(lambda x: ko.pos(x, join=True))
```

```python
sentiword[0] = sentiword[0].apply(lambda x: ';'.join(x))
```

```python
sentiword = sentiword[43:]
```

```python
sentiword = sentiword.rename(columns={0:'ngram', 1:'score'})
```

```python
inter = list(set(polarity['ngram']).intersection(set(sentiword['ngram'])))
```

```python
for i in inter:
    print(sentiword.drop(sentiword.loc[sentiword['ngram'] == i].index, inplace=True))
```
